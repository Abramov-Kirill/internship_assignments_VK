```python
# Посмотрим на распределение значений в каждом столбце при замене выбросов на медианное значение
pd.options.mode.chained_assignment = None
def replace_numerical_outliers(column_name, z_thresh=3):
  median = column_name.median()
  std = column_name.std()
  outliers = ((column_name - median).abs()) > z_thresh*std
  column_name[outliers] = np.nan
  column_name.fillna(median, inplace=True)
  return column_name



for col in train_df_c.loc[:, train_df_c.columns != 'target']:
    train_df_c[col] =  replace_numerical_outliers(train_df_c[col])
```

# LGBMRanker

```python
def predict_model(model, test_x, test_q):
  predictions = []

  for group in np.unique(test_q):
    preds = model.predict(test_x[test_q == group])
    predictions.extend(preds)

  return predictions


def ndcg(y_score, y_true, k):
  order = np.argsort(y_score)[::-1]
  y_true = np.take(y_true, order[:k])

  gain = 2 ** y_true - 1

  discounts = np.log2(np.arange(len(y_true)) + 2)
  return np.sum(gain / discounts)

def mean_ndcg(model, test_x, test_y, test_q):
  qids = np.unique(test_q)
  ndcg_ = list()

  for i, qid in enumerate(qids):
    y = test_y[test_q == qid]
    y = np.array(y).reshape(-1)
    if np.sum(y) == 0:
        continue

    p = model.predict(test_x[test_q == qid])

    idcg = ndcg(y, y, k=10)
    ndcg_.append(ndcg(p, y, k=10) / idcg)

  return np.mean(ndcg_)
```

```python
from lightgbm import LGBMRanker

# Строим модель LGBMRanker, подбираем параметры и считаем NDCG

boosting_types = ['gbdt', 'dart']
num_leaves = [10, 20, 30, 40, 50, 60]
subsamples_for_bin = [100, 200, 400, 500, 700, 1000, 2000, 3000, 5000, 10000]


best_mean_ndcg = 0
best_para = []
for b_t in boosting_types:
  for n_l in num_leaves:
    for s_f_b in subsamples_for_bin:

      gbm = LGBMRanker(boosting_type=b_t, num_leaves=n_l, subsample_for_bin=s_f_b)
      gbm.fit(train_x, train_y, group=group_train)

      mean_ndcg_t = mean_ndcg(gbm, test_x, test_y, test_q)
      if mean_ndcg_t > best_mean_ndcg:
        best_mean_ndcg = mean_ndcg_t
        best_param = [b_t, n_l, s_f_b]
print(best_mean_ndcg)
print(best_param)
```

Результаты NDCG по LGBMRanker равны  0.6717333342720152

# XGBRanker

```python
train_x = train_df.loc[:, ~train_df.columns.isin(['target'])]
train_q = train_df['search_id']
train_y = train_df['target']

test_q = test_df['search_id']
```

```python
from sklearn.model_selection import LeaveOneGroupOut

train_data= train_df.copy()
X_train = train_data.loc[:, ~train_data.columns.isin(['search_id','target'])]
y_train = train_data.loc[:, train_data.columns.isin(['target'])]

groups = train_data.groupby('search_id').size().to_frame('size')['size'].to_numpy()

test_df_c = test_df.copy()
test_df_c = test_df.loc[:, train_df.columns]

test_x = test_df_c.loc[:, ~test_df_c.columns.isin(['search_id', 'target'])]
test_y = test_df_c.loc[:, test_df_c.columns.isin(['target'])]

_, group_train = np.unique(train_q, return_counts=True)
```

```python
import xgboost as xgb
from tqdm import tqdm

# Строим модель XGBRanker, подбираем параметры и считаем NDCG

boosters = ['gbtree', 'gblinear']
max_depths = [5,6,7,8,9,10, 15, 20]
n_estimators = [30, 50, 70, 90, 110, 130, 150]


best_mean_ndcg = 0
best_para = []

for b in boosters:
  for m_d in max_depths:
    for n_e in n_estimators:
      model = xgb.XGBRanker(
        booster=b,
        learning_rate=0.1,
        max_depth=m_d,
        n_estimators=n_e
      )
      model.fit(X_train, y_train, group=groups, verbose=True)

      mean_ndcg_t = mean_ndcg(model, test_x, test_y, test_q)
      if mean_ndcg_t > best_mean_ndcg:
        best_mean_ndcg = mean_ndcg_t
        best_param = [b, m_d, n_e]
print(best_mean_ndcg)
print(best_param)
```

Результаты NDCG по XGBRanker равны 0.6265792246192613
